## Section 2: Malthusian Model

The first model we examine assumes that the population density simply changes by a fixed factor of $r>0$ ($r$ is called the **reproduction rate**) per day. In symbols, this mathematical relationship reads
$$
p_{n+1} = \left(r+1\right)p_{n}. 
$$
We call this model the **Malthusian model**, since it is related to Thomas Malthus' early work in population dynamics. 

Note how our model allows us to compute $p_{n}$ for any $n\geq 0$ knowing only $p_0$ (which we would just take from the data), by recursively applying the growth formula above. Don't underestimate the importance of this property: it is the ecological, discrete-time version of Newton's conception of dynamics (compute particle trajectories knowing only their initial states and the forces they experience)! In mathematical terms, this means we're modelling the population as a **discrete-time dynamical system** (dynamical=time-varying). More specifically, one calls our update rule a **difference equation** since it can be written as condition on the difference of the densities at successive times. 

By recursively applying the Malthusian update rule to an initial state $p_0$, we find that for any $n>0$ we have
$$
p_n = \left(1+r\right)^{n} p_0. 
$$
So, as long as $r>0$, $p_n$ grows without bound as $n$ gets large. 

Now, let's simulate the Malthusian model for various values of $r$ to get a feeling for what it does. 

```python
def simulate_malthus(r, p0, T):
    
    p_malthus = np.zeros(T-1, dtype=float) # initialize the output

    p_malthus[0] = p0 # take initial state from the data

    for n in range(1,T-1):# T-1 bcz we start at time 2
        
        p_malthus[n] = (r+1) * p_malthus[n-1]
        
    return p_malthus
```

Let's just pick some value for $r>1$ and see what the simulation spits out up to some time $T$. Soon, we'll have more to say about how to pick $r$ based on the data. 

```python
r = 1.

T = int(5) # start by picking a small $T$, then gradually increasing it. 

p_malthus = simulate_malthus(r, 14., T)
```

```python
fig, ax =plt.subplots()   
    
plt.plot(np.arange(2,T+1), p_malthus, color='xkcd:reddish purple', linestyle='dashed', label='Malthus', linewidth=2)
plt.plot(np.arange(2,T+1), dens[:T-1], 'o-', color='xkcd:greenish', label='Data')

plt.xlabel(r"$n$", fontsize=26, color = 'k')
plt.ylabel(r"$p_{n}$", fontsize=26, color='k')

plt.tick_params(axis='x', which='both', top='off', color = 'k')
plt.xticks(fontsize=16, rotation=0, color = 'k')
plt.tick_params(axis='y', which='both', right='off', color = 'k')
plt.yticks(fontsize=16, rotation=0, color = 'k')
ax.locator_params(axis='y', nbins=6)
ax.locator_params(axis='x', nbins=4)

ax.legend(fontsize=16)

plt.xlim([2, T])
plt.ylim([0, 1.05*np.amax(dens[:T-1])])

plt.frameon = True 
plt.grid(visible=True)

plt.tight_layout()

fig.set_size_inches(8, 6)  
        
#figname = 'malthus_sim.png'
#plt.savefig(figname, dpi=600)

plt.show()
```

By playing around with the simulation length $T$ (default is $T=5$, and it can be increased up to $T=25$) it looks like the Malthusian model qualitatively captures the initial rapid growth in the density, but fails to capture the level-off. 

Why should this be "in plain English"? Well, the Malthusian model assumes the population always increases at a fixed rate. Is this biologically reasonable? Perhaps, if the supply of nutrients for the paramecia increased with the population, but it doesn't: Gause kept the level of nutrients in his population constant. So, while it is reasonable to expect fast, Malthusian growth for a short time, eventually there are so many organisms that they must start competing with one another for the fixed amount of food available. This competition will slow the growth rate down: based on our data, the growth rate should in fact tend to about zero. So, a constant growth rate over the entire experiment period must be thrown out. 

All in all, it seems that the Malthus model is a decent fit over short times, but a pretty bad fit for longer times, and common-sense scientific reasoning would lead us to expect this anyway. 

However, to be $100 \%$ sure that the Malthusian model can't be saved, we should take some time to compute the optimal value of the parameter $r$. Ok, how do we do this? The answer is to use the data at hand. 

By subtracting $p_{n}$ from both sides of the Malthusian recurrence relation, we find
$$
\Delta p_{n} = rp_{n}. 
$$
Thus, $r$ is simply the slope of the line of best fit on a $\Delta p_{n}$ versus $p_{n}$ graph. So, all we have to do is get that line of best fit! Though the underlying theory is a bit beyond the scope of this course, finding a line of best fit is probably one of the easiest coding tasks there is, and it can be solved a bunch of different ways. For the benefit of your Python knowledge, I'll show you two different ways using SciPy and SciKit-Learn. We'll see that both ways produce the same estimate for $r$. 

The code below runs for the same simulation length $T$ used above, so for small $T$ you should see a good fit but as $T$ increases up to $25$ you should see the fit getting worse and worse. 

```python
# For convenience when making plots later, sort data before we do regression. 
# Neither regressor cares about the sorting, but it does make our lives easier. 
# I got this solution from the discussion here:
# https://stackoverflow.com/questions/32019398/python-sorting-y-value-array-according-to-ascending-x-array

import operator

def data_for_fitting(T):
    
    x = dens[:T-1]
    y = dens_ch[:T-1]

    L = sorted(zip(x,y), key=operator.itemgetter(0))

    x, y = zip(*L)

    x = np.array(x)
    y = np.array(y)
    
    return x,y
```

```python
x,y = data_for_fitting(T)

#### scipy version
from scipy.stats import linregress

sp_regression = linregress(x,y) # NOTE: have to cut off the manipulated variable!

r_best = sp_regression.slope

print('SciPy Regressor Estimate = ', r_best)
```

```python
#### sklearn version for comparison.... 
from sklearn import linear_model

sklearn_regressor = linear_model.LinearRegression()
xlen = np.shape(dens[:T-1])[0]
sklearn_regressor.fit(x.reshape(xlen,1), y)

r_best_sklearn = sklearn_regressor.coef_

print('SciKit-Learn Regressor Estimate = ', r_best_sklearn)
```

We see that there is no difference between the two different slope estimates. For instance, they both give
$$
r = 0.971757
$$
when $T=5$ and
$$
r = -0.132293
$$
where $T=25$. The latter result should activate your applied mathematician's spidey sense: if $r<0$, the Malthusian model predicts *shrinking*, not growth, at every time step! So, parameter "optimization" for large times is producing an even worse fit! 

You may ask, then, why we have bothered with such an optimization here (or really why we've bothered with Malthus in the first place). The answer is that this model is so simple that it makes for an ideal playground in which to explain parameter fitting and modelling validation (more like invalidation!): if you master the "how" for the Malthusian model, understanding the more complicated model-testing in subsequent sections will be easy. Besides, we have confirmed at least that Malthus' model works well for small time implying that, whatever large time model we use, its short-time behaviour must resemble Malthusian growth.  

Now, we plot the regressed line against the actual data. 

```python
fig, ax = plt.subplots()   
    
plt.plot(x, sp_regression.slope*x+sp_regression.intercept, color='xkcd:soft purple', linestyle='dashed', label='Best Fit Line', linewidth=2)
plt.plot(x,y, '^-', color='xkcd:burnt red', label='Data', linewidth=2)

plt.xlabel(r"$p_{n}$", fontsize=26, color = 'k')
plt.ylabel(r"$\Delta p_{n}$", fontsize=26, color='k')

plt.tick_params(axis='x', which='both', top='off', color = 'k')
plt.xticks(fontsize=16, rotation=0, color = 'k')
plt.tick_params(axis='y', which='both', right='off', color = 'k')
plt.yticks(fontsize=16, rotation=0, color = 'k')
ax.locator_params(axis='y', nbins=12)
ax.locator_params(axis='x', nbins=8)

ax.legend(fontsize=16)

plt.xlim([0.99*np.amin(x), 1.01*np.amax(x)])
if np.amin(y) < 0 : 
    plt.ylim([1.01*np.amin(y),1.1*np.amax(y)])
else: plt.ylim([0.25*np.amin(y),1.1*np.amax(y)])
    
plt.frameon = True 
plt.grid(visible=True)

plt.tight_layout()

fig.set_size_inches(8, 6)  
        
#figname = 'malthus_regression.png'
#plt.savefig(figname, dpi=600)

plt.show()
```

Qualitatively, the fit is pretty poor for large times, but does it at least produce a good model? For large times, we already know the answer is "no" since $r<0$, but let's simulate the Malthus model so we can really wallow in its awfulness:

## Section 3: Discrete Logistic Model

The big issue with the Malthusian model was that it implicitly assumed the paramecia had infinitely many resources on which to feed, and thereby sustain a constant growth rate. Of course, since the paramecia actually have a finite, constant amount of available nutrients, the growth rate needs to eventually tend to zero as the population gets large. In biology, this is called **logistic growth**. 

Let's try to model a density-dependent growth law $f(p_n)$ that gives rise to (approximately) logistic growth. That is, we want to look for models that are discrete dynamical systems of the form 
$$
\Delta p_{n} = f(p_n).
$$

What constraints must $f(p_n)$ satisfy? Here are some ideas:  

1) First, $f(0)=0$, since organisms cannot magically appear out of nowhere. Note that this is automatically satisfied if $f(p) = pr(p)$, which is often the case in practice. 

2) Second, $f(p_{n})$ should vanish when $p_{n}$ reaches a certain threshold density $K>0$: $r(K)=0$. This would represent the growth rate going to zero when the population gets sufficienty big. 

3) Finally, we need $f(p_n)\geq 0$ to guarantee growth.

The simplest function satisfying both constraints involves two parameters $a,K>0$: 

$$
f(p_{n}) = a(K-p_{n})p_{n}. 
$$

This leads to a model of the form 

$$
p_{n+1} = p_{n} + a(K-p_{n})p_{n},
$$

called the **discrete logistic model (DLM)**. Let's start experimenting with this model using Python. 

We use nonlinear fitting with SciPy to estimate the best parameters from $a,K$. This is like finding the *parabola* of best fit on the $\Delta p_n$ vs. $p_{n}$ data, rather than the line of best fit like we saw with Malthus. Don't worry about the "why" here, I just want to show you that Python packages can be used to perform the important scientific task of parameter estimation. 

```python
from scipy.optimize import curve_fit

T = 25

def parabola(x, a, K):
    return a*(K-x)*x

x,y = data_for_fitting(T)

params_opt = curve_fit(parabola, x,y)
 
a_best = params_opt[0][0]

K_best = params_opt[0][1]

print('Optimal value of a = %.5f' % a_best)
print('Optimal value of K = %.1f' % K_best)
```

Now, we write a function for simulating DLM up to a prescribed time $T$. 

```python
def simulate_dlm(a,K,p0,T):
    
    p_dlm = np.zeros(T-1, dtype=float)

    p_dlm[0] = p0 # take initial state from the data

    for n in range(1,T-1):
        
        p_dlm[n] = p_dlm[n-1] + a * p_dlm[n-1] * (K-p_dlm[n-1])
        
    return p_dlm
```

We simulate DLM with the optimal parameters and plot the results and Gause's data on the same axes. 

```python
p_dlm_best = simulate_dlm(a_best, K_best,14., T)
```

```python
fig, ax =plt.subplots()   

plt.plot(np.arange(2,T+1), p_dlm_best, color='xkcd:blueberry', linestyle='dotted', label='DLM', linewidth=2)
plt.plot(np.arange(2,T+1), dens, 'o-', color='xkcd:greenish', label='Data')

plt.xlabel(r"$n$", fontsize=26, color = 'k')
plt.ylabel(r"$p_{n}$", fontsize=26, color='k')

plt.tick_params(axis='x', which='both', top='off', color = 'k')
plt.xticks(fontsize=16, rotation=0, color = 'k')
plt.tick_params(axis='y', which='both', right='off', color = 'k')
plt.yticks(fontsize=16, rotation=0, color = 'k')
ax.locator_params(axis='y', nbins=12)
ax.locator_params(axis='x', nbins=8)

ax.legend(fontsize=16)

plt.xlim([2, 24])
plt.ylim([0, 1.05*np.amax(dens)])

plt.frameon = True 
plt.grid(visible=True)

plt.tight_layout()

fig.set_size_inches(8, 6)  
        
#figname = 'dlm_sim_best.png'
#plt.savefig(figname, dpi=600)

plt.show()
```

Wow! This fit looks pretty good. So, even a simple method that accounts for finite nutrients vastly improves model quality. 

You may think that we can summarize the quality of the fit by reporting the mean-square error of "data - fit". However, this isn't really scientifically responsible, as the fit has been *trained* on the data! It's best statistical practice to either 

1) perform a "test-train split" where you train the fit only on part of the data, then evaluate its quality on the remaining data

2) perform another identical experiment of *P. aurelia* growth, get some new data, and then report the mean-square error of "new data - old fit".

Since our data set is too small to perform a meaningful test-train split and we unfortunately don't have access to a microbiology lab for this course, both options are unfortunately off the table for us. Nevertheless, for your future studies keep in mind the difference between using data to *obtain* a model and using data to *validate* a model. 

Now comes the bad news. Despite the DLM providing an excellent fit, it has some features that make it a less-than-desirable model. For instance, depending on the values of $a, K$, it can actually exhibit **chaotic** behaviour, which for now we'll understand to simply mean non-periodic "oscillations". Here is an example of a chaotic trajectory for DLM:

## Section 3: Beverton-Holt Model

Now, we present a third model, the **Beverton-Holt model (BH)** that has all the best parts of DLM, but avoids the nasty chaos. We'll simply write the model down for now and verify that is satisfies the same three constraints that we placed on DLM. 

In terms of two parameters $b, L$, the dynamical update rule for BH reads 

$$
p_{n+1} = \frac{b}{1+\frac{b-1}{L}p_{n}} p_{n}. 
$$

which can be re-arranged to yield

$$
\Delta p_{n} = p_{n} \left( \frac{b}{1+\frac{b-1}{L}p_{n}} -1 \right),
$$

the form of the update rule we use for parameter estimation via nonlinear regression. 

1) First, we immediately verify that $f(p_n)$ is zero when $p_{n}=0$. 

2) Second, a bit of algebra shows that the growth rate also vanishes if $p_{n}=L$. So, the growth rate has a root at some positive threshold value. 

3) Guaranteeing a positive growth rate is a little tougher because the parameters are so complicated, so we won't check this until we actually perform simulations

First, let's find the optimal parameters $b,L$ with nonlinear fitting, as we did with DLM. 

```python
def bh_curve(x, b, L):
    return x*(-1. + b*L/(L+(b-1.)*x))

bh_params_opt = curve_fit(bh_curve, x, y, p0=[1., K_best], bounds= [[1, 0], [np.inf, np.inf]])
# for some reason the optimizer has a tough time with the BH curve, probably because it has 
# issues around L=0. So, you have to put bounds in to get a reasonable result! 
    
b_best = bh_params_opt[0][0]

L_best = bh_params_opt[0][1]

print('Optimal value of b = %.5f' % b_best)
print('Optimal value of L = %.1f' % L_best)
```

Now, we create a function for simulating BH. 

```python
def simulate_bh(b,L,p0,T):
    
    p_bh = np.zeros(T-1, dtype=float)

    p_bh[0] = p0 # take initial state from the data

    for n in range(1,T-1):
        
        p_bh[n] = p_bh[n-1]*(b*L)/(L+(b-1)*p_bh[n-1])
        
    return p_bh
```

We simulate BH with the optimal parameters. 

```python
p_bh_best = simulate_bh(b_best, L_best ,14., T)
```

```python
fig, ax =plt.subplots()   

plt.plot(np.arange(2,T+1), p_bh_best, linestyle='dashdot', color='xkcd:rust', label='BH', linewidth=2)
plt.plot(np.arange(2,T+1), dens, 'o-', color='xkcd:greenish', label='Data')

plt.xlabel(r"$n$", fontsize=26, color = 'k')
plt.ylabel(r"$p_{n}$", fontsize=26, color='k')

plt.tick_params(axis='x', which='both', top='off', color = 'k')
plt.xticks(fontsize=16, rotation=0, color = 'k')
plt.tick_params(axis='y', which='both', right='off', color = 'k')
plt.yticks(fontsize=16, rotation=0, color = 'k')
ax.locator_params(axis='y', nbins=12)
ax.locator_params(axis='x', nbins=8)

ax.legend(fontsize=16)

plt.xlim([2, 24])
plt.ylim([0, 1.05*np.amax(dens)])

plt.frameon = True 
plt.grid(visible=True)

plt.tight_layout()

fig.set_size_inches(8, 6)  
        
#figname = 'bh_sim_best.png'
#plt.savefig(figname, dpi=600)

plt.show()
```

BH produces a population curve that roughly fits the data nicely.

Additionally, while DLM may have chaotic solutions, BH has the following desirable property: 

**Theorem**: As $n\rightarrow \infty$, solutions of BH with $p_0>0$ always tend to

$$
p_{\text{eq}} = \frac{bL}{b-1}. 
$$

That is, $p_{\text{eq}}$ is a **stable equilibrium**. 

*Sketch of Proof*

Theorems of this sort are usually proven with stability and bifurcation analysis and other tools from dynamical systems theory. However, for BH, the proof is very easy: one simply verifies via the substitution $p_{n}\mapsto u_{n} = p_{n}^{-1}$ that the exact solution for $p_{n}$ (evolving according to BH) is 

$$
p_{n} = \frac{b^{n}}{1+\frac{b^{n}-1}{L}p_0}p_{0}. 
$$

You can check this yourself if you know how to solve linear difference equations (for an intro to this topic, see *An Introduction to Difference Equations* ed.3 by Saber Elaydi). 

Having an exact solution for a dynamical system is very rare, so don't count on it in general! With this solution in hand, we simply compute the limit assuming $p_0\neq 0$:

\begin{align*}
\lim_{n\rightarrow \infty} p_{n} &= bL \lim_{n\rightarrow \infty} \frac{b^{n-1}}{(b^{n}-1)+\frac{L}{p_0}}
\\ \\ 
&= bL \lim_{n\rightarrow \infty} \frac{b^{n-1}}{(b-1)(b^{n-1} + b^{n-2}+\cdots+b+1)+\frac{L}{p_0}}
\\ \\ 
&= p_{\text{eq}} \lim_{n\rightarrow \infty} \frac{b^{n-1}}{(b^{n-1} + b^{n-2}+\cdots+b+1)+\frac{L}{p_0(b-1)}}
\\ \\ 
&= p_{\text{eq}},
\end{align*}
so we're all done. 


This means that Beverton-Holt indeed describes logistic growth: it fits the data and is backed up by solid mathematics. So, out of the three dynamical models considered so far, Beverton-Holt seems to give the best overall model for Gause's laboratory observations. 

## Section 4: Ricker Model

Next, we discuss the *Ricker model* of population growth. Like DLM, the Ricker model can exhibit chaos for certain parameter values, so it's not quite as useful as BH. Nevertheless, the Ricker equation is an important standard model in ecology (including fisheries modelling, see for instance *Quantitative Fish Dynamics* by Quinn and Deriso), so it's a good idea for us to cover it for the sake of completeness alone. Additionally, studying this model gives us a good opportunity to review the all-important **exponential function**. 

In terms of two parameters $c, M$, the dynamical update rule for Ricker reads 

$$
p_{n+1} = \exp\left(c\left(1-\frac{p_n}{M}\right)\right) p_{n}. 
$$

which can be re-arranged to yield

$$
\Delta p_{n} = p_{n} \left[\exp\left(c\left(1-\frac{p_n}{M}\right)\right) - 1\right].
$$

Here, $\exp(x) = e^{x}$ denotes the **exponential function**, where $e$ is the irrational number $e = 2.71828...$ . The following expressions for the exponential function are sometimes useful: 

**Theorem** 

For any $x\in \mathbb{R}$, the exponential function $\exp(x)$ admits the following representations: 

1) $\exp(x) = \lim_{n\rightarrow\infty} \left(1+\frac{x}{n}\right)^{n},$ and

2) $\exp(x) = \sum_{k=0}^{\infty} \frac{x^k}{k!} = 1 + x + \frac12 x^2 +\frac16 x^3 + \cdots $ . 

In particular, $\exp(0) = e^0 = 1$, $e^{x}\rightarrow 0$ as $x\rightarrow \infty$, and $e^{-x}\rightarrow 0$ as $x\rightarrow \infty$.

As we've done many times before, we simulate Ricker dynamics and plot the results against the given data curve. 

```python
def simulate_ricker(c,M,p0,T):
    
    p_ricker = np.zeros(T-1, dtype=float)

    p_ricker[0] = p0 # take initial state from the data

    for n in range(1,T-1):
        
        p_ricker[n] = p_ricker[n-1]*np.exp(c*(1.-p_ricker[n-1]/M))
        
    return p_ricker
```
